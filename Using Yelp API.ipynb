```python
#Imports
from yelpapi import YelpAPI
import json
import pandas as pd
```

```python
#Open API credentials
with open('/Users/krisbarber/.secret/yelp_api.json') as f:
    login = json.load(f)

login.keys()
```

```python
#Create API connection
yelp_api = YelpAPI(login['api-key'], timeout_s = 5.0)
yelp_api
```

```python
#Create API call for Italian food in Denver
italian = yelp_api.search_query(location='Denver, CO',
                                       term='Italian')
print(type(italian))
italian.keys()
```

```python
#Create dataframe from API results
italian = pd.DataFrame(italian['businesses'])
italian.head(2)
```

```python
#Find the first 2 pages of results
italian_20 = yelp_api.search_query(location='Denver, CO',
                                       term='Italian',
                               offset = 20)
```

```python
#Create second dataframe
italian20 = pd.DataFrame(italian_20['businesses'])
italian20.head(2)
```

```python
#Concatenate into one dataframe
all_italian = pd.concat([italian, italian20],
                      ignore_index=True)
display(all_italian.head(3), all_italian.tail(3))
```

```python
#Save results to JSON file
all_italian.to_json(orient='records')
```
